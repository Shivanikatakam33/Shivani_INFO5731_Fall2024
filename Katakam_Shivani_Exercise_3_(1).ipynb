```python
Sentiment Analysis is one of the interesting text classification task. It is commonly used for examining customer reviews, survey responses and social media comments.
We can perform sentiment analysis on customer reviews so as to classify the reviews as positive, negative and neutral.
The main applications of sentiment analysis could be in the business sector for understanding customer reviews, delivering better services by improving the products.
The dataset will be a huge collection of customer reviews from an ecommerce platform.
In order to build a machine learning model,choosing relevant features is very important.
The five different types of features are:
1. Bag of Words:
Here, the text is represented as unordered collection of words. the words resembles features and the value represents the frequency of the words.
This is mainly used to represent the presence or absence of special words in the text. For example words like Good, Beautiful represents the positivity where as words like poor, dissapointed indicates negativity.
2. Term Frequency-Inverse Document Frequency:
This is used to estimate the significance of a word in a document. Popular words like the are penalized, and less common but significant terms are given greater weight.
Words that are more specific to a given review can be highlighted with the use of TF-IDF. Reducing noise from popular words that don't really help identify emotion can help models perform better.
3. Sentiment Scores:
Texts can be scored for polarity that is positive, neutral, or negative using sentiment lexicons like TextBlob or VADER.
It is possible to add these scores as features.  Sentiment scores offer a high-level overview of the review's tone and are closely linked to the sentiment classification task.
4. Part of Speech Tags:
POS tagging assigns a grammatical category such as noun, verb, or adjective to each word in a phrase.
Various speaking patterns can shed light on the sentiment. Sentiment analysis relies heavily on adjectives and adverbs such as good or horribly to convey feelings.
5. N-grams Bigrams or Trigrams:
Continuous sequences of 'n' items from a given text sample are called N-grams.
These elements could be words in this instance trigrams are three-word phrases, bigrams are two-word phrases.
Word dependencies that single-word features like BoW cannot capture are captured by n-grams. For example, not good is a negative term, while good by itself could be viewed as positive.
Such complex concepts can be better conveyed with bigrams and trigrams.


```

## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk import pos_tag, word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfTransformer
from nltk.sentiment import SentimentIntensityAnalyzer
from collections import Counter
import nltk
texts = [
    "I am loving this poduct! It's beautiful and works great.",
    "This product has very bad quality. So disappointed.",
    "This is the worst purchase i have ever made, it didn't meet my expectations.",
]
bow_vectorizer = CountVectorizer()
bow_features = bow_vectorizer.fit_transform(texts)
print("Bag of Words Features:\n", pd.DataFrame(bow_features.toarray(), columns=bow_vectorizer.get_feature_names_out()))
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(texts)
print("\nTF-IDF Features:\n", pd.DataFrame(tfidf_features.toarray(), columns=tfidf_vectorizer.get_feature_names_out()))
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
sentiment_scores = [sia.polarity_scores(text) for text in texts]
print("\nSentiment Scores:\n", pd.DataFrame(sentiment_scores))
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
def get_pos_tags(text):
    tokens = word_tokenize(text)
    return pos_tag(tokens)
pos_tags = [get_pos_tags(text) for text in texts]
print("\nPOS Tags:\n", pos_tags)
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
bigrams = bigram_vectorizer.fit_transform(texts)
print("\nBigrams:\n", pd.DataFrame(bigrams.toarray(), columns=bigram_vectorizer.get_feature_names_out()))



```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
from sklearn.feature_selection import chi2
import numpy as np
y = np.array([1, 0, 0])
bow_features = bow_vectorizer.fit_transform(texts)
# Compute Chi-Square values
chi2_scores, p_values = chi2(bow_features, y)
# DataFrame with their Chi-Square scores and sort them
feature_names = bow_vectorizer.get_feature_names_out()
chi2_df = pd.DataFrame({'Feature': feature_names, 'Chi2 Score': chi2_scores, 'p-value': p_values})
# Sort features by Chi-Square score in descending order
ranked_features = chi2_df.sort_values(by='Chi2 Score', ascending=False)
# ranked features
print("Ranked Features based on Chi-Square scores:\n", ranked_features)




```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
from transformers import BertTokenizer, BertModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze()
# Sample text
texts = [
    "I am loving this poduct! It's beautiful and works great.",
    "This product has very bad quality. So disappointed.",
    "This is the worst purchase i ahve ever made, it didn't meet my expectations."
]
# query
query = "The product works great and I love it."
# embeddings for the texts and the query
text_embeddings = [get_bert_embedding(text) for text in texts]
query_embedding = get_bert_embedding(query)
# cosine similarity between the query and each text
similarities = [cosine_similarity(query_embedding.unsqueeze(0), text_embedding.unsqueeze(0)).item() for text_embedding in text_embeddings]
# Rank texts by similarity score in descending order
ranked_texts = sorted(zip(texts, similarities), key=lambda x: x[1], reverse=True)
for i, (text, sim) in enumerate(ranked_texts):
    print(f"Rank {i+1}: (Similarity: {sim:.4f}) - Text: {text}")




```
